# Model Selection

### Otimization
Each model was tuned with optuna for the same task.

We have a wide variety of models including linear(logreg), probabilistic(naive_bayes), tree based(random forest),neaural network(MLP) and tree based boosting(xgb).

### Evaluation and comparison

```python
# Importing models with optimized hyperparameters
rf_model = joblib.load('saved_models/random_forest_model.pkl')
logreg_model = joblib.load('saved_models/logistic_regression_model.pkl')
svm_model = joblib.load('saved_models/svc_model.pkl')
xgb_model = joblib.load('saved_models/xgb_model.pkl')
mlp_model = joblib.load('saved_models/mlp_model.pkl')
gnb_model = joblib.load('saved_models/naive_bayes_model.pkl')

models=[rf_model,logreg_model,svm_model,xgb_model,mlp_model,gnb_model]
model_names = ['Random Forest','Logistic Regression', 'SVC', 'XGBoost', 'MLP', 'Naive Bayes']
```

```python
# Define features and target
X = df_churn.drop(columns=['Churn'])
y = df_churn['Churn']

# Split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)

# Scale features for models that need scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python

# Evaluation metrics that wil help choosing the best model
results = pd.DataFrame(columns=['Model', 'Mean Accuracy', 'Mean Precision', 'Mean Recall', 'Mean F1 Score', 'Mean ROC AUC', 
                                'Best Threshold', 'Accuracy After Threshold', 'Precision After Threshold', 'Recall After Threshold', 
                                'F1 Score After Threshold', 'ROC AUC After Threshold'])

# Thresholds to loop over 
thresholds_range = np.arange(0.0, 1.01, 0.01) 

# Evaluate each model
for model, name in zip(models, model_names):
    if name in ['Logistic Regression', 'SVC', 'MLP']:
        X_data = X_train_scaled
    else:
        X_data = X_train

    # Cross-validate model
    cv_accuracy = cross_val_score(model, X_data, y_train, cv=5, scoring='accuracy')
    cv_precision = cross_val_score(model, X_data, y_train, cv=5, scoring='precision')
    cv_recall = cross_val_score(model, X_data, y_train, cv=5, scoring='recall')
    cv_f1 = cross_val_score(model, X_data, y_train, cv=5, scoring='f1')
    cv_roc_auc = cross_val_score(model, X_data, y_train, cv=5, scoring='roc_auc')

    # Fit the model on the full training data to get probabilities
    model.fit(X_data, y_train)
    y_probs = model.predict_proba(X_data)[:, 1]  

    # Metrics for manual thresholding
    best_f1_score = -1
    best_threshold = None

    # Loop through the thresholds to find the best one
    for threshold in thresholds_range:
        # Apply the threshold to get binary predictions
        y_pred_thresholded = (y_probs >= threshold).astype(int)

        # Calculate metrics after applying the threshold
        precision_after_threshold = precision_score(y_train, y_pred_thresholded, zero_division=1)
        recall_after_threshold = recall_score(y_train, y_pred_thresholded, zero_division=1)
        f1_after_threshold = f1_score(y_train, y_pred_thresholded, zero_division=1)

        # Keep track of the threshold with the best F1 score
        if f1_after_threshold > best_f1_score:
            best_f1_score = f1_after_threshold
            best_threshold = threshold
            best_precision_after_threshold = precision_after_threshold
            best_recall_after_threshold = recall_after_threshold
            best_accuracy_after_threshold = accuracy_score(y_train, y_pred_thresholded)

    # Calculate ROC AUC on the probabilities, not affected by threshold
    roc_auc_after_threshold = roc_auc_score(y_train, y_probs)

    # Store mean metrics from cross-validation and the best thresholding metrics
    new_result = pd.DataFrame({
        'Model': [name],
        'Mean Accuracy': [cv_accuracy.mean()],
        'Mean Precision': [cv_precision.mean()],
        'Mean Recall': [cv_recall.mean()],
        'Mean F1 Score': [cv_f1.mean()],
        'Mean ROC AUC': [cv_roc_auc.mean()],
        'Best Threshold': [best_threshold],
        'Accuracy After Threshold': [best_accuracy_after_threshold],
        'Precision After Threshold': [best_precision_after_threshold],
        'Recall After Threshold': [best_recall_after_threshold],
        'F1 Score After Threshold': [best_f1_score],
        'ROC AUC After Threshold': [roc_auc_after_threshold]
    })

    # Concatenate the result
    results = pd.concat([results, new_result])
    
results = results.sort_values(by='Mean ROC AUC', ascending=False)
results
```

- Mean Accuracy: Shows how often the model predicts correctly. NOTE accuracy can be misleading if the data is imbalanced and this is the case for our and most of the churn datasets.
- Mean Precision: The proportion of true churn predictions out of all churn predictions made. High precision means fewer false positives.
- Mean Recall: Measures how well the model identifies actual churn cases. High recall means fewer false negatives.
This metric can be very important from a business perspective, high FN can generate a lot of cost for the company as they are missing the opportunity to reach out and try to retain the customers.
- Mean F1 Score: Balance between precision and recall.
- ROC AUC: The ability of the model to distinguish between churn and non-churn. Higher ROC AUC is better, as it shows a better trade-off between true positive and false positive rates.
This is imortant, because we want to minimize False negatives(missed churners), but also want to keep False positives low to avoid a high nuber of 'false alarms' as it can generate unnecessary cost for the company.
Worth to mention here that to find the right balance we need to now how high is the average cost of a churn and the cost of retaining customers.
- Post-Threshold Metrics: After applying the "Best Threshold," these metrics show how well the model performs when adjusted for class imbalance or business requirements.

```python
# predictions with the best threshold
for model, name in zip(models, model_names):
    if name in ['Logistic Regression', 'SVC', 'MLP']:
        model.fit(X_train_scaled, y_train)
        y_probs = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_probs = model.predict_proba(X_test)[:, 1]

    # Apply the best threshold
    y_pred_adjusted = (y_probs >= results.loc[results['Model'] == name, 'Best Threshold'].values[0]).astype(int)

    # Confusion matrix and classification report
    conf_matrix = confusion_matrix(y_test, y_pred_adjusted)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title(f'Confusion Matrix for {name} with Threshold Tuning')
    plt.show()

    class_report = classification_report(y_test, y_pred_adjusted)
    print(f'Classification Report for {name} with Threshold Tuning:\n')
    print(class_report)
```

### OPTION 2:

    Based on the probabilities we can group customers by their churn risk.
    This can help the company to create different strategies for customers with different levels of churn risk.

```python
# Method 1: Using Probability Thresholds
def assign_risk_levels_thresholds(probas, thresholds):
    risk_levels = []
    for proba in probas:
        if proba < thresholds[0]:
            risk_levels.append('Low')
        elif proba < thresholds[1]:
            risk_levels.append('Medium')
        elif proba < thresholds[2]:
            risk_levels.append('High')
        else:
            risk_levels.append('Very High')
    return risk_levels

thresholds = [0.25, 0.5, 0.75]  
churn_with_probas['risk_level'] = assign_risk_levels_thresholds(churn_with_probas['xgb_proba'], thresholds)

# Method 2: Using Probability Quartiles
def assign_risk_levels_quartiles(probas):
    quartiles = np.percentile(probas, [0, 25, 50, 75, 100])
    risk_levels = []
    for proba in probas:
        if proba <= quartiles[1]:
            risk_levels.append('Low')
        elif proba <= quartiles[2]:
            risk_levels.append('Medium')
        elif proba <= quartiles[3]:
            risk_levels.append('High')
        else:
            risk_levels.append('Very High')
    return risk_levels

churn_with_probas['risk_level_quartile'] = assign_risk_levels_quartiles(churn_with_probas['xgb_proba'])
```

```python
churn_with_probas.head()
```


- First Method: Manual Probability Thresholds
In this approach, we set specific cut-off points for the predicted churn probabilities to decide risk levels. For example, if a customer's churn probability is higher than 0.75, they will be labeled as "Very High" risk. 

- Second Method: Quartile Ranges
This method uses the distribution of all predicted probabilities to assign risk levels. We look at the top 25% of predicted probabilities and classify those customers as "Very High" risk. This segmentation is more flexible because it adapts to the data distribution.


The main difference between these two methods is how they decide on risk levels. The first method uses fixed numbers to classify customers, while the second method uses the actual spread of probabilities in the data. Because of this, the two methods can give different risk levels to customers, depending on how the probabilities are distributed.

```python

# Counts of each risk level for Method 1
risk_counts_method1 = churn_with_probas['risk_level'].value_counts().reset_index()
risk_counts_method1.columns = ['Risk Level', 'Count']  
risk_counts_method1['Method'] = 'Probability Thresholds' 

# Counts of each risk level for Method 2
risk_counts_method2 = churn_with_probas['risk_level_quartile'].value_counts().reset_index()
risk_counts_method2.columns = ['Risk Level', 'Count']  
risk_counts_method2['Method'] = 'Quartiles'  

# Combine both counts 
risk_counts = pd.concat([risk_counts_method1, risk_counts_method2], ignore_index=True)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(
    data=risk_counts, 
    x='Risk Level', 
    y='Count', 
    hue='Method', 
    ax=axes[0],  
    palette='Set2'
)

axes[0].set_title('Risk Levels Comparison')
axes[0].set_xlabel('Risk Level')
axes[0].set_ylabel('Customer Count')

sns.histplot(
    data=churn_with_probas, 
    x='xgb_proba', 
    kde=True, 
    bins=20,  
    ax=axes[1],  
    color='steelblue'
)

axes[1].set_title('Predicted Churn Probability Distribution')
axes[1].set_xlabel('Predicted Churn Probability')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
